Group data using custom function: Let's say you want to group your data using custom function. Here the requirement is to create three groups

Q-1) Days when temperature was between 80 and 90

Q-2) Days when it was between 50 and 60

Q-3) Days when it was anything else

```python
df
```

```python
def group (x):
    if 80<x<90:
        return"80-90"
    elif 50<x<60:
        return "50-60"
    else:
        return "others"
```

```python
df["new"]=df["temperature"].apply(group)
df.groupby("new").max()
```

```python
df.loc[(df["temperature"]<90) & (df["temperature"]>80),["temperature"]]
```

# MERGE FUNCTION

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```

```python

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python

df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3
```

```python

df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python

df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```
