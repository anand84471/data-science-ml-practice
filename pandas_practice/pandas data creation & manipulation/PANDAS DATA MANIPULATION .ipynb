# Diff B/W MAP, APPLY , APPLY MAP & TRANSFORM

```python
df = pd.DataFrame({'AGE': [20,21,32,25,30], 'MARKS': [100,40,60,90,64] })
df
```

```python
def sqrt(x):
    return pow(x,0.5)
```

```python
df.apply(sqrt)
```

```python
df["AGE"].map(sqrt)
```

```python
df.transform(sqrt)
```

```python
df.apply(lambda x: pow(x,0.5))
```

```python
df.transform(lambda x: pow(x,0.5))
```

```python
df["AGE"].map(lambda x: pow(x,0.5))
```

```python
df['age_sqrt'] = df['AGE'].apply(sqrt)
df
```

```python

df['marks_lambda'] = df['MARKS'].apply(lambda x: x-20)
df
```

```python
df['AGE_SQRT_T'] = df['AGE'].transform(sqrt)
df
```

```python
df['MARKS_LAMB_TRA'] = df['MARKS'].transform(lambda x: x+10)
df
```

```python

df['AGE_MAP'] = df['AGE'].map(sqrt)
df
```

```python

df["MARKS_MAP_LAMB"] = df['MARKS'].transform(lambda x: x/10)
df
```

```python
df.transform({'AGE': np.sqrt,'MARKS': np.exp,})
```

```python
df.apply({'AGE': np.sqrt,'MARKS': np.exp,})
```

```python
df.apply(lambda x:x.sum())
```

```python
df.transform(lambda x:x.sum())
```

```python
df.apply(lambda x: x['AGE'] - x['MARKS'],axis=1)
```

```python
df.transform(lambda x: x['AGE'] - x['MARKS'],axis=1)
```
