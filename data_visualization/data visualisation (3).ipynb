```python
wer=data.select_dtypes("object")
wer
```

```python
wer1=data.drop(wer,axis=1)
wer1
```

```python
col=data.columns
col
col1=['mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc']
col1
```

```python
a=dict(alpha=.6,bins=15,edgecolor="g",color="r")
for i in wer1:
    plt.figure(figsize=(11,4))
    plt.hist(data[i],**a)
    plt.title("histogram of {}".format(i))
    plt.xlabel("{}".format(i))
    plt.ylabel("frequency")
    plt.show()
    
    
   
```

```python
plt.figure(figsize=(25,10))
col=["r","y","c","olive","orange","peru","b","g"]
index=1
for i in wer1:
    if index<=8:
        ax=plt.subplot(2,4,index)
        plt.hist(data[i],label=i,bins=20,color=col[index-1],edgecolor="k")
        plt.title("HISTOGRAM OF {}".format(i))
        plt.xlabel(i)
        plt.ylabel("FREQUENCY OF {}".format(i))
        plt.legend(loc="upper right")
        ax.axis("tight")
        index+=1
```

```python
plt.figure(figsize=(21,15))
a=dict(bins=20,edgecolor="r",color="g")
index=1
for i in col1:
    if index<=8:
        ax=plt.subplot(3,3,index)
        plt.hist(data[i],**a,label=i)
        plt.title("HISTOGRAM OF {}".format(i))
        plt.xlabel(i)
        plt.ylabel("FREQUENCY OF {}".format(i))
        plt.legend(loc="upper right")
        ax.axis("tight")
        index+=1
```
