# Visualizing statistical relationships

Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

### 1. Numerical Data Ploting
- relplot()
- scatterplot()
- lineplot()

### 2. Categorical Data Ploting
- catplot()
- boxplot()
- stripplot()
- swarmplot()
- etc...

### 3. Visualizing Distribution of the Data
- distplot()
- kdeplot()
- jointplot()
- rugplot()

### 4. Linear Regression and Relationship
- regplot()
- lmplot()


```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Seaborn has five built-in themes to style its plots: darkgrid, whitegrid, dark, white, and ticks

```python
sns.set(style = 'darkgrid')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python

sns.relplot(x ='total_bill', y = 'tip', data = tips,height=4,aspect=3)
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', style = 'size', data = tips,height=5,aspect=3)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips,height=4,aspect=3)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips,style="sex",hue="size",height=5,aspect=3)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips,hue = 'smoker', style = 'time', size = 'size',height=5,aspect=3)
```

```python

sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'sex', data = tips,height=5,aspect=2)
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', size = 'size', style = 'time')
```

```python
#catplot()
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,height=5,aspect=2.5)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, jitter = False,height=5,aspect=2.5)
```

```python
sns.catplot(x = 'day', y = 'tip', data = tips, kind = 'swarm',aspect=2.5,hue="sex")
```

```python

sns.catplot(x = 'day', y = 'tip', data = tips,order=["Sun","Sat","Fri","Thur"],height=5,aspect=2.5)
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x = 'cut', y = 'price', kind = 'box',data=diamonds,height=5,aspect=2.5)
```

```python
sns.catplot(x = 'color', y = 'price', kind = 'boxen', data = diamonds,height=5,aspect=2.5)
```

```python
sns.catplot(y= 'total_bill', x = 'day', hue = 'sex', kind = 'violin', data = tips,height=5,aspect=2.5)
```

```python
g = sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', inner = None, data = tips,height=5,aspect=2.5)
sns.swarmplot(x = 'day', y = 'total_bill', color = 'k', size = 3, data = tips, ax = g.ax)
```
