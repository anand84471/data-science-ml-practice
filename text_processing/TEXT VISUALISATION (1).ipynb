```python
p4 = re.sub('<[^<]+?>','', p3)
```

```python
p5 = ''.join(c for c in p4 if not c.isdigit())
```

# ALTERNATIVE WAY

```python
paragraph
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences[2]
```

```python
corpus=[]
for i in range(len(sentences)):
    review=re.sub("[^a-zA-Z]"," ",sentences[i])
    review=review.lower()
    review=review.split()
    review=[ls.lemmatize(j) for j in review if not j in stopwords.words("english")]
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
alt_p7=" ".join([i for i in corpus])

```
