# 1) One Hot Encoding

```python
df_ob_ohc=pd.get_dummies(df_ob,drop_first=True)
df_ob_ohc.head()
```

```python
df_ob["JobType"]=pd.get_dummies(df_ob["JobType"],drop_first=True)
```

# 2) Take top 10 categories from a column (merely for nominal feature)

```python
df_ob_10=df_ob.copy()
```

```python
df_ob_10.shape
```

```python
for i in df_ob.columns:
    print(i,len(df_ob[i].unique()))
```

```python
df_ob_10["nativecountry"].value_counts().head().index
```

```python
class_10=[x for x in df_ob_10["nativecountry"].value_counts().head(10).index]

for i in class_10:
    df_ob_10[i]=np.where(df_ob_10["nativecountry"]== i,1,0)    
```

```python
df_ob_10.head()
```

# 4) Target Guided Ordinal Encoding ( merely for Classification Models)

```python
df_ob.head()
```

```python
df_tgor=df_ob.copy()
```

```python
df_tgor["SalStat"]=pd.get_dummies(df_tgor["SalStat"],drop_first=True)
```

```python
df_tgor.head()
```

```python
for i in df_ob.columns:
    print(i,len(df_ob[i].unique()))
```

```python
df_tgor["JobType"].unique()
```

```python
df_tgor["JobType"]=df_tgor["JobType"].replace(' ?',df_tgor["JobType"].mode()[0])
```

```python
df_tgor.groupby(['JobType'])['SalStat'].mean()
```

```python
df_tgor.groupby(['JobType'])['SalStat'].mean().sort_values(ascending=False).index
```

```python
tgor=df_tgor.groupby(['JobType'])['SalStat'].mean().sort_values(ascending=False).index
```

```python
ordinal_labels2={k:i for i,k in enumerate(tgor,0)}
ordinal_labels2
```

```python
df_tgor['job_tgor']=df_tgor['JobType'].map(ordinal_labels2)
df_tgor.head()
```

# 5) Mean Encoding

```python
mean_ordinal=df_tgor.groupby(['occupation'])['SalStat'].mean().to_dict()
```

```python
df_tgor["occupation"]=df_tgor["occupation"].replace(' ?',df_tgor["occupation"].mode()[0])
```

```python
mean_ordinal
```

```python
df_tgor['occupation_me']=df_tgor['occupation'].map(mean_ordinal)
df_tgor.head()
```
